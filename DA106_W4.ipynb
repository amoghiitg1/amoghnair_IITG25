```python
#importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
# Loading the Iris dataset from seaborn
df = sns.load_dataset('iris')
```

```python
# Loading a preview of the dataset
df.head(10)
```

```python
# Plotting the graphs - Histogram
plt.figure(figsize=(10,6))
sns.histplot(data=df.sepal_length, bins=30)
plt.title("Histogram of Sepal Length")
plt.show
```

```python
# Plotting the graphs - Density plot
plt.figure(figsize=(10,6))
sns.kdeplot(data=df.sepal_length)
plt.title("Density plot of Sepal Length")
plt.show
```

```python
# Plotting the graphs - Density plot
plt.figure(figsize=(10,6))
sns.kdeplot(data=df.sepal_length)
sns.rugplot(data=df.sepal_length)
plt.title("Density plot_Rug plot of Sepal Length")
plt.show
```

```python
# Plotting the graphs - Box plot
plt.figure(figsize=(10,6))
sns.boxplot(x="species", y="sepal_length", data=df)
plt.title("Box plot of Sepal Length")
plt.show
```

```python
# Plotting the graphs - Correlation map
plt.figure(figsize=(10,6))
sns.heatmap(df.iloc[:, 0:4].corr(),annot=True, cmap="coolwarm")
plt.title("Density plot of Sepal Length")
plt.show
```

```python
# Creating a dataframe with some missing values
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan]
})
```

```python
df
```

```python
# create an instance of simpleimputer with mean strategy
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns = df.columns)
df_imputed
```
